```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Netflix average movie duration data (2011–2020)
netflix_dict = {
    "Year": [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    "Average Duration (min)": [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]
}
```

```python
netflix_df = pd.DataFrame(netflix_dict)
```

```python
netflix_df
```

```python
min_duration = netflix_df["Average Duration (min)"].min()
max_duration = netflix_df["Average Duration (min)"].max()

print("Minimum duration:", min_duration)
print("Maximum duration:", max_duration)
```

```python
plt.plot(netflix_df["Year"], netflix_df["Average Duration (min)"], marker='o', color='blue')
plt.title("Netflix Movie Durations (2011–2020)")
plt.xlabel("Year")
plt.ylabel("Average Duration (min)")
plt.grid(True)
plt.show()
```

```python
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head()
```

```python
netflix_movies_df = netflix_df[netflix_df["type"] == "Movie"]
```

```python
netflix_movies_df.head()
```

```python
movies_subset = netflix_movies_df[["release_year", "duration"]]
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.scatter(movies_subset["release_year"], movies_subset["duration"], color="red", alpha=0.5)
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release Year")
plt.ylabel("Duration (min)")
plt.show()
```
